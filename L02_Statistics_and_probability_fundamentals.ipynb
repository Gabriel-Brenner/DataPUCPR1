**Bland-Altman plot**

A [Bland-Altman plot](https://en.wikipedia.org/wiki/Bland%E2%80%93Altman_plot), also known as a Tukey mean-difference plot or a Bland-Altman diagram, is a statistical method used to assess the agreement between two quantitative measurements.

**References**

**Bland J.M., Altman, D.G. (1986)**. [Statistical methods for assessing agreement between two methods of clinical measurement](https://pubmed.ncbi.nlm.nih.gov/2868172/), Lancet, vol. 327, no. 8476, pp. 307–310.

**Bland J.M., Altman, D.G. (1999)**. [Measuring agreement in method comparison studies](https://pubmed.ncbi.nlm.nih.gov/10501650/), Statistical Methods in Medical Research. vol. 8, no. 2, pp. 135–160.

```python
import matplotlib.pyplot as plt
import numpy as np

def bland_altman_plot(data1, data2, *args, **kwargs):
    data1 = np.asarray(data1)
    data2 = np.asarray(data2)
    mean  = np.mean([data1, data2], axis=0)
    diff  = data1 - data2                    # Difference between data1 and data2
    md    = np.mean(diff)                    # Mean of the difference
    sd    = np.std(diff, axis=0)             # Standard deviation of the difference

    plt.scatter(mean, diff, *args, **kwargs)
    plt.axhline(md, color='gray', linestyle='--')
    plt.axhline(md + 1.96*sd, color='gray', linestyle='--')
    plt.axhline(md - 1.96*sd, color='gray', linestyle='--')

    # Annotate the lines with their values
    plt.text(plt.xlim()[1], md, f'mean {md:.2f}', va='center', ha='left', color='green')
    plt.text(plt.xlim()[1], md + 1.96*sd, f'std {md + 1.96*sd:.2f}', va='center', ha='left', color='green')
    plt.text(plt.xlim()[1], md - 1.96*sd, f'std {md - 1.96*sd:.2f}', va='center', ha='left', color='green')

# Example usage:
data1 = np.random.normal(10, 2, 100)
data2 = np.random.normal(10, 2, 100)

bland_altman_plot(data1, data2)
plt.show()
```

## <font color='green'> **SM: 2.3 Statistical measures**

* **Mean**: the average value of the variable. You sum up all the values and divide by the number of values.

* **Median**: the middle value in the dataset when it's sorted in ascending order. If there's an even number of data points, it's the average of the two middle values.

* **Minimum**: the smallest value of the variable.

* **Maximum**: the largest value of the variable.

* **Standard deviation, $\sigma$**: measures the dispersion or spread of the variable. It indicates how much the values differ from the mean value.

* **Skewness**: It measures the asymmetry of the variable's distribution. Positive skewness means the data is skewed to the right, negative skewness means it's skewed to the left, and zero skewness indicates a symmetric distribution.

* **Kurtosis**: It is a measure of the tailedness of a distribution. Tailedness is how often outliers occur. Excess kurtosis is the tailedness of a distribution relative to a normal distribution.

* Distributions with medium kurtosis (medium tails) are mesokurtic.

* Distributions with low kurtosis (thin tails) are platykurtic.

* Distributions with high kurtosis (fat tails) are leptokurtic.

* **Q1 (First quartile)**: It is the value below which 25% of the data falls.

* **Q3 (Third quartile)**: It is the value below which 75% of the data falls.

* **Interquartile Range (IQR)**: It is the range between the first quartile (Q1) and the third quartile (Q3). It gives a measure of the spread of the middle 50% of the data.

* **Coefficient of Variation (CV)**: It is the ratio of the standard deviation to the mean. It's a measure of relative variability and is useful for comparing variability between datasets with different units or scales. If the mean is zero, it's often considered undefined or reported as NaN (not a number) to indicate that it's not meaningful in that context.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, skew, kurtosis

print('Gaussian signal analysis\n')

# Generate Gaussian signal and calculate statistics
sigma_values = range(1, 11)  # sigma values from 1 to 10
statistics   = []

for sigma in sigma_values:
    x        = np.linspace(-10, 10, 1000)  # x-axis values
    signal   = norm.pdf(x, loc=0, scale=sigma)  # Gaussian signal
    mean     = np.mean(signal)
    median   = np.median(signal)
    min_val  = np.min(signal)
    max_val  = np.max(signal)
    std      = np.std(signal)
    skewness = skew(signal)
    kurt     = kurtosis(signal)
    q1       = np.percentile(signal, 25)
    q3       = np.percentile(signal, 75)
    iqr      = q3 - q1                                  # Interquartile Range
    cv       = std / mean if mean != 0 else np.nan      # Coefficient of Variation
    statistics.append((sigma, mean, median, min_val, max_val, std, skewness, kurt, q1, q3, iqr, cv))

# Print statistics
print("{:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10} {:<10}".format("Sigma", "Mean", "Median",
                                                                                                   "Min", "Max", "Std", "Skewness", "Kurtosis", "Q1", "Q3", "IQR", "CV"))
print("-" * 140)
for stat in statistics:
    print("{:<10} {:<10.5f} {:<10.5f} {:<10.5f} {:<10.5f} {:<10.5f} {:<10.5f} {:<10.5f} {:<10.5f} {:<10.5f} {:<10.5f} {:<10.5f}".format(*stat))
print("\n")

# Plot Gaussian signals for different standard deviations
plt.figure(figsize=(10, 6))

for sigma, mean, median, min_val, max_val, std, skewness, kurt, _, _, _, _ in statistics:
    x      = np.linspace(-10, 10, 1000)       # x-axis values
    signal = norm.pdf(x, loc=0, scale=sigma)  # Gaussian signal
    plt.plot(x, signal, label=f'$\sigma$ = {sigma}')

plt.title('Gaussian signal for different standard deviations, $\sigma$')
plt.xlabel('$x$')
plt.ylabel('Probability density')
plt.legend()
#plt.grid(True)
plt.show()
```

## <font color='green'> **SM: 2.4 Statistical measures - Boxplot visualization example**

A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset.

It displays the five-number summary of the data:
* minimum
* first quartile (Q1)
* median (second quartile or Q2)
* third quartile (Q3)
* maximum.

image.png

In a boxplot, the **whiskers** are the lines that extend from the box to indicate the variability outside the upper and lower quartiles.

They represent the range of the data distribution beyond the interquartile range (IQR).

[Source](https://www.geeksforgeeks.org/interquartile-range-and-quartile-deviation-using-numpy-and-scipy/)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate Gaussian signal and calculate statistics
sigma_values = range(1, 11)  # sigma values from 1 to 10
signals = []

for sigma in sigma_values:
    x      = np.linspace(-10, 10, 1000)  # x-axis values
    signal = norm.pdf(x, loc=0, scale=sigma)  # Gaussian signal
    signals.append(signal)

# Calculate quartiles for each signal
q1_values = [np.percentile(signal, 25) for signal in signals]
q3_values = [np.percentile(signal, 75) for signal in signals]

# Plot boxplot with interquartile range
plt.figure(figsize=(10, 6))
plt.boxplot(signals, positions=sigma_values, widths=0.4, patch_artist=True, medianprops=dict(color='black'))

# Draw the interquartile range (IQR)
for i in range(len(sigma_values)):
    plt.plot([sigma_values[i] - 0.15, sigma_values[i] + 0.15], [q1_values[i], q1_values[i]], color='gray')
    plt.plot([sigma_values[i] - 0.15, sigma_values[i] + 0.15], [q3_values[i], q3_values[i]], color='gray')
    plt.plot([sigma_values[i], sigma_values[i]], [q1_values[i], q3_values[i]], color='black')

plt.title('Boxplot of Gaussian signal with interquartile range (IQR)')
plt.xlabel('Standard deviation, $\sigma$')
plt.ylabel('Probability density')
plt.grid(True)
plt.show()
```

## <font color='green'> **SM: 2.5 Probability density function**

A probability density function (PDF) is a function used in probability theory and statistics to describe the likelihood of a continuous random variable taking on a specific value within a given range.

Unlike the probability mass function (PMF) used for discrete random variables, which assigns probabilities to specific outcomes, the PDF describes the relative likelihood of different values occurring within a continuous range.

The **Kolmogorov-Smirnov (KS) test** is a statistical method used to determine whether a sample distribution differs significantly from a reference probability distribution (often a standard normal distribution or another specified theoretical distribution).

It is particularly useful when you want to assess if two samples are drawn from the same underlying distribution.

**References**

**Kolmogorov, A. (1933)**. Sulla determinazione empirica di una legge di distribuzione. Giornale dell'Istituto Italiano degli Attuari, 4, 83-91. (Title translated to English: "On the empirical determination of a distribution law")

**Smirnov, N. (1936)**. Sur la distribution de ω2 (criterium de M. Kolmogoroff). Comptes Rendus de l'Académie des Sciences de l'URSS, 15, 353-356. (Title translated to English: "On the distribution of ω2 (criterion of M. Kolmogorov)")

```python
%%time
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from scipy.stats import (norm, gamma, expon, lognorm, beta, t, logistic, gumbel_r, gumbel_l, genextreme,
                         uniform, pareto, pearson3, ks_2samp)
from sklearn.metrics import mean_squared_error

# Load the dataset: Iris dataset
iris = datasets.load_iris()
data = iris.data[:, 0]  # Selecting just one feature for simplicity

# Plot histogram of the data
plt.figure(figsize=(16, 12))

# Fit and plot various probability density functions
pdfs = {
    "Normal"             : norm,
    "Gamma"              : gamma,
    "Exponential"        : expon,
    "Lognormal"          : lognorm,
    "Beta"               : beta,
    "Student-t"          : t,
    "Logistic"           : logistic,
    "Gumbel Right"       : gumbel_r,
    "Gumbel Left"        : gumbel_l,
    "Uniform"            : uniform,
    "Generalized Pareto" : pareto,
    "Pearson Type III"   : pearson3
}

mse_results = {}
ks_results  = {}

# Setting up the subplot grid
rows = 4
cols = 3
fig, axes = plt.subplots(rows, cols, figsize=(8, 6))

for idx, (name, pdf) in enumerate(pdfs.items()):
    if pdf is None:
        mse_results[name] = np.inf
        ks_results[name] = (np.nan, np.nan)
        continue

    params = pdf.fit(data)

    x = np.linspace(min(data), max(data), 100)
    pdf_values = pdf.pdf(x, *params)

    # Plot histogram of the data
    ax = axes[idx // cols, idx % cols]
    ax.hist(data, bins=30, density=True, alpha=0.6, color='g')

    # Plot the PDF
    ax.plot(x, pdf_values, label=name)
    ax.set_title(name)

    # Calculate MSE
    bin_centers, _ = np.histogram(data, bins=30)
    pdf_data = pdf.pdf((bins[:-1] + bins[1:]) / 2, *params)
    mse = mean_squared_error(bin_centers, pdf_data)
    mse_results[name] = mse

    # Kolmogorov-Smirnov (KS) test
    rv = pdf(*params)
    ks_stat, p_value = ks_2samp(data, rv.rvs(size=len(data)))
    ks_results[name] = (ks_stat, p_value)

# Adjust layout
plt.tight_layout()

# Print ranking of best results in terms of MSE
sorted_results = sorted(mse_results.items(), key=lambda x: x[1])
print("Ranking of PDFs based on MSE (Mean Squared Error):\n")

print("{:<25} {:<15}".format("Probability Distribution", "MSE"))
print("-" * 40)
for idx, (pdf_name, mse) in enumerate(sorted_results, start=1):
    print("{:<25} {:<15.5f}".format(pdf_name, mse))
print("\n")

# Print KS Test results
print("\nKolmogorov-Smirnov (KS) test results:\n")
print("{:<25} {:<20} {:<20}".format("Probability distribution", "KS statistic", "p-value"))
print("-" * 65)
for pdf_name, (ks_stat, p_value) in ks_results.items():
    print("{:<25} {:<20.5f} {:<20.5f}".format(pdf_name, ks_stat, p_value))
print("\n")
plt.show()
```

**Brief explanation of the pdfs**

* **Normal distribution**: It is characterized by a bell-shaped curve and is symmetric around its mean.

* **Gamma distribution**: A continuous probability distribution with two parameters, often used to model waiting times or durations.

* **Exponential distribution**: Describes the time between events in a Poisson point process, with a constant event rate.

* **Lognormal distribution**: Results from the exponential of a normally distributed random variable, often used to model data that is skewed to the right.

* **Beta distribution**: Defined on the interval [0, 1], often used to model proportions or probabilities.

* **Student-t distribution**: Similar to the normal distribution but with heavier tails, useful when dealing with small sample sizes or when data may have outliers.

* **Logistic distribution**: Symmetric S-shaped distribution often used in modeling growth processes.

* **Gumbel right distribution**: Represents the maximum of a number of samples of various distributions, commonly used in extreme value theory.

* **Gumbel left distribution**: Similar to the Gumbel Right distribution but for the minimum instead of the maximum.

* **Uniform distribution**: All outcomes are equally likely, with a constant probability.

* **Generalized Pareto distribution**: Used to model the tails of another distribution, often used in extreme value theory.

* **Pearson type III distribution**: A three-parameter family of distributions that includes several common distributions such as the normal, gamma, and exponential distributions.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

import kalepy as kale

NUM = int(1e4)
np.random.seed(100)

# Combine data from two different PDFs
_d1  = np.random.normal(4.0, 1.0, NUM)
_d2  = np.random.lognormal(0, 0.5, size=NUM)
data = np.concatenate([_d1, _d2])

# Calculate the "true" distribution
xx  = np.linspace(0.0, 7.0, 100)[1:]
yy  = 0.5*np.exp(-(xx - 4.0)**2/2) / np.sqrt(2*np.pi)
yy += 0.5 * np.exp(-np.log(xx)**2/(2*0.5**2)) / (0.5*xx*np.sqrt(2*np.pi))
```

```python
# Plot histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=50, density=True, alpha=0.7, color='blue', label='Data histogram')

# Plot true distribution
plt.plot(xx, yy, 'r--', linewidth=2, label='True distribution')

# Calculate and annotate metrics
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
plt.axvline(mean, color='k', linestyle='dashed', linewidth=1, label='Mean')
plt.axvline(median, color='g', linestyle='dashed', linewidth=1, label='Median')

# Add legend and labels
plt.legend()
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Combined distribution')

# Show plot
plt.show()

# Print metrics
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Standard deviation: {std_dev:.2f}")
```

Reconstruct the probability-density based on the given data points.

```python
points, density = kale.density(data, probability=True)

# Plot the PDF
plt.plot(points, density, 'k-', lw=2.0, alpha=0.8, label='KDE')

# Plot the "true" PDF
plt.plot(xx, yy, 'r--', alpha=0.4, lw=3.0, label='truth')

# Plot the standard, histogram density estimate
plt.hist(data, density=True, histtype='step', lw=2.0, alpha=0.5, label='hist')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Kernel density estimation (KDE)')

plt.legend()
```

Draw new samples from the KDE reconstructed PDF.

```python
samples = kale.resample(data)

# Plot new samples
plt.hist(samples, density=True, label='new samples', alpha=0.5, color='0.65', edgecolor='b')
# Plot the old samples
plt.hist(data, density=True, histtype='step', lw=2.0, alpha=0.5, color='r', label='input data')

# Plot the KDE reconstructed PDF
plt.plot(points, density, 'k-', lw=2.0, alpha=0.8, label='KDE')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Kernel density estimation (KDE)')

plt.legend()
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KernelDensity

# Fit Kernel Density Estimate (KDE) to the data
kde = KernelDensity(kernel='gaussian', bandwidth=0.5).fit(data.reshape(-1, 1))

# Draw new samples from the KDE reconstructed PDF
samples = kde.sample(len(data))

# Plot new samples
plt.hist(samples, density=True, label='New Samples', alpha=0.5, color='0.65', edgecolor='b')

# Plot the original input data
plt.hist(data, density=True, histtype='step', lw=2.0, alpha=0.5, color='r', label='Input Data')

# Evaluate KDE on a grid of points
points = np.linspace(min(data), max(data), 1000)
density = np.exp(kde.score_samples(points.reshape(-1, 1)))

# Plot the KDE reconstructed PDF
plt.plot(points, density, 'k-', lw=2.0, alpha=0.8, label='KDE')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Kernel density estimation')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
