# **Data quality**

Each aspect of data quality plays a critical role in ensuring that data is reliable, meaningful, and useful for data science purposes.

* **Accuracy**: Accuracy refers to <font color='red'> how closely the data reflects the true <font color='balck'> value or state of the entity or phenomenon it represents.

* **Completeness**: Completeness indicates whether <font color='red'>all necessary data<font color='black'> is present in a dataset.

* **Consistency**: Consistency refers to the <font color='red'>uniformity and coherence of data<font color='black'> across different sources or within the same dataset.

* **Timeliness**: Timeliness reflects how up-to-date the data is relative to the <font color='red'>timeframe of analysis or decision-making.<font color='black'>

* **Relevance**: Relevance pertains to the degree to which the data is pertinent or <font color='red'>applicable to the specific context <font color='black'> or problem being addressed.

* **Validity**: Valid data accurately captures the intended meaning or <font color='red'>characteristics without distortion or bias<font color='black'>.

* **Uniformity**: Uniformity relates to the <font color='red'>consistency of data format<font color='black'>, structure, and representation across the dataset.

# **Some potentialities of statistical methods in data science**

* **Descriptive statistics**: Statistical methods allow for summarizing and describing the main features of a dataset. Measures like <font color='red'> mean, median, mode, variance, and standard deviation<font color='black'> provide insights into the central tendency, variability, and distribution of data.

* **Inferential statistics**: With inferential statistics, data scientists can infer <font color='red'>properties of a population based on a sample.<font color='black'> Techniques such as hypothesis testing, confidence intervals, and regression analysis help make predictions and draw conclusions from data.

* **Predictive modeling**: Statistical methods enable the development of predictive models to forecast future trends or outcomes based on historical data. Techniques like <font color='red'>regression analysis, time series analysis, and machine learning algorithms <font color='black'>leverage statistical principles to make accurate predictions.

* **Experimental design**: Statistical methods guide the design and analysis of experiments, allowing researchers to systematically manipulate variables and assess the impact on outcomes. <font color='red'>Design of Experiments (DOE) techniques <font color='black'> optimize experimentation processes and improve the efficiency of data collection.

* **Pattern recognition**: Statistical methods aid in identifying patterns and relationships within data, helping to uncover insights and trends that may not be immediately apparent. <font color='red'>Clustering techniques and dimensionality reduction methods <font color='black'> are examples of statistical approaches for pattern recognition.

* **Anomaly detection**: Statistical methods can identify <font color='red'> outliers or anomalies <font color='black'> in data that deviate from expected patterns. These anomalies may represent errors, fraud, or significant events requiring further investigation.

* **Bayesian inference**: Bayesian statistical methods offer a probabilistic framework for updating beliefs in light of new evidence. Bayesian techniques are particularly useful in <font color='red'>situations with limited data or when incorporating prior knowledge into the analysis.<font color='balck'>

* **Causal inference**: Statistical methods help assess <font color='red'>causal relationships between variables<font color='black'>, allowing data scientists to determine the impact of one variable on another while controlling for confounding factors.

* **Validation and model evaluation**: Statistical methods provide tools for validating models and evaluating their performance. Techniques such as <font color='red'>cross-validation and bootstrapping <font color='black'> assess the accuracy and robustness of predictive models.

## <font color='green'> **SM: 4.1 Imputation of missing values**

Handling missing data is a crucial step in data preprocessing, ensuring robust analyses and accurate model training. Here are some common methods for imputing missing values:

1. **Mean/Median imputation**: It involves substituting missing values with either the <font color='red'>**mean or median** <font color='black'> of the existing data. It's particularly effective when the absence of data occurs randomly and when there's a limited number of missing data points.

2. **Arbitrary value imputation**: It involves substituting missing values with predetermined arbitrary values.  <font color='red'>**Tree-based models**<font color='black'> excel in managing this type of imputation due to their ability to adapt to non-linear relationships and handle categorical variables effectively.

3. **Frequent category imputation**: It is a handy technique for handling <font color='red'>missing categorical data<font color='black'> by replacing them with the <font color='red'>**mode**<font color='black'>, or most frequently occurring category. This method shines when missingness occurs randomly and when there are relatively few missing data points.

4. **Random sample imputation**: This method involves substituting missing data with a <font color='red'>**randomly** selected value from the existing dataset<font color='black'>. Unlike other techniques, it maintains the variable's distribution without distortion. However, it introduces randomness into the feature engineering process, which may affect model stability and interpretability.

5. **Multivariate imputation**: It involves leveraging <font color='red'>**regression models to forecast** absent values<font color='black'>. This approach offers a refined estimation of missing data, aiming to reduce distortion in variables. However, its integration into the machine learning pipeline introduces heightened complexity.

**Pattern of missing data**

**Classifications** used to describe the patterns of missing data in a dataset:

* **MCAR** (**M**issing **C**ompletely **a**t **R**andom): Signifies that the <font color='red'>missing data occurs randomly <font color='black'> across the dataset, independent of any observed or unobserved variables.

* **MAR** (**M**issing **a**t **R**andom): Indicates that the likelihood of <font color='red'>missing data is related solely to the observed data<font color='black'>, meaning the probability of data being missing is determined by the values of other observed variables.

* **MNAR** (**M**issing **N**ot **a**t **R**andom): Suggests that the probability of <font color='red'>missing data is influenced by unobserved factors or variables that are not captured within the dataset<font color='black'>, leading to a systematic bias in the missing data pattern.

**k-Nearest Neighbors (KNN) imputation**

KNN imputation is a technique used to fill in missing values in a dataset based on the values of neighboring data points.

The choice of K (the number of nearest neighbors) can impact the imputation results and may require tuning based on the specific dataset and problem at hand.

```python
from sklearn.impute import SimpleImputer
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

# Sample data with missing values
X = [[1, 2, None],
     [3, None, 4],
     [5, 6, 7]]

# 1. Imputation under MCAR (Mean/Median Imputation)
# For MCAR, we'll use SimpleImputer with strategy='mean' or 'median'
imputer_mcar = SimpleImputer(strategy='mean')  # or strategy='median'
X_imputed_mcar = imputer_mcar.fit_transform(X)

print("Imputed under MCAR (Mean/Median Imputation):")
print(X_imputed_mcar)

# 2. Imputation under MAR (Multiple Imputation)
# For MAR, we'll use IterativeImputer
imputer_mar = IterativeImputer()
X_imputed_mar = imputer_mar.fit_transform(X)

print("\nImputed under MAR (Multiple Imputation):")
print(X_imputed_mar)

# 3. Imputation under MNAR (Not handled by sklearn directly, but we can simulate)
# For MNAR, we'll simulate a scenario and impute using a custom approach
import numpy as np

# Simulating MNAR scenario where missingness depends on the missing values themselves
X_mnar = np.array([[1, 2, np.nan],
                   [3, np.nan, 4],
                   [5, 6, np.nan]])

# Define a custom imputation strategy for MNAR (e.g., replacing missing values with -1)
X_imputed_mnar = np.where(np.isnan(X_mnar), -1, X_mnar)

print("\nImputed under MNAR (Custom Approach):")
print(X_imputed_mnar)

```

* For **MCAR**, it usea `SimpleImputer` with the strategy set to '**mean**' or '**median**'.

* For **MAR**, it usea `IterativeImputer` which performs multiple imputation based on observed variables.

* For **MNAR**, it simulates a scenario where missingness depends on the missing values themselves and replace missing values with a custom approach. MNAR is not directly handled by `scikit-learn`, so we use a custom strategy.

## <font color='green'> **SM: 4.2 Outlier identification and removal**

Anomaly detection refers to a group of machine learning techniques whose aim is to spot anomalies in a set of observations, usually to halt unwanted behaviors in your systems.

The term anomaly is often used as a synonym for outlier. There is, however, a subtle but crucial difference between the two.

* Outlier identification and removal is a crucial step in data analysis and statistics.

* Outliers are data points that deviate significantly from the rest of the data in a dataset.

* They can arise due to various reasons such as **measurement errors, data entry mistakes, or genuinely unusual observations**.

----

* An **outlier** <font color='green'> is an event that, given a certain data distribution, is unlikely to happen, a low probability occurrence<font color='black'>.

* An **anomaly** <font color='red'> is an event that is virtually impossible to explain, given the same data distribution<font color='balck'>.

----

While outliers are accidental and can be caused by chance, anomalies have deeper roots and can be traced back to causal conditions or events. And to the extent that there are specific mechanisms that generate an anomaly, the job of anomaly detection, especially in a business context, is to discover them.


Identifying and removing outliers is important because they can skew statistical analyses and machine learning models, leading to inaccurate results and conclusions. Outliers can influence the mean and standard deviation of a dataset, affecting measures of central tendency and variability. They can also distort relationships between variables in regression analysis.

**Outlier and anomaly**

Outlier and anomaly are two terms commonly used in data analysis and statistics, but they have slightly different meanings and contexts:

image.png

**Outlier**:

* An outlier is a data point that significantly differs from other observations in a dataset. It is often an extreme value that lies outside the typical range of the majority of the data.
* Outliers can occur due to various reasons, including measurement errors, natural variability, or genuine but rare events.
* Identifying outliers is crucial in data analysis because they can distort statistical analyses and modeling results if not properly handled. Outliers may need to be investigated to determine if they are valid data points or if they should be treated differently in the analysis.
Anomaly:

**Anomaly**:

* An anomaly is a deviation or departure from what is considered normal or expected in a dataset. It is a broader term than outlier and can encompass various types of unexpected behavior or events.
* Anomalies can be outliers, but they can also include patterns, trends, or events that are unusual or unexpected but not necessarily extreme in value.
* Anomalies can be detected through various techniques, including statistical methods, machine learning algorithms, or domain-specific knowledge.

**Summary**:

While outliers are specific extreme values in a dataset, anomalies refer to any deviations or unexpected patterns, which may include outliers but extend to other types of irregularities as well.
